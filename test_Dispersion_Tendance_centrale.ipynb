```python
# Importation des bibliothèques nécessaires
#!pip install tabulate
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
# Chargement du dataset Iris
iris = sns.load_dataset('iris')

# Affichage des premières lignes du dataset
print(iris.head())
```

```python
iris.dtypes
```

```python
# Moyenne
mean = iris.mean(numeric_only=True)
print("Moyenne:\n", mean)

# Médiane
median = iris.median(numeric_only=True)
print("\nMédiane:\n", median)

# Mode
mode = iris.mode(numeric_only=True).iloc[0]
print("\nMode:\n", mode)

# Calcul de la variance pour chaque variable
variance = iris.var(numeric_only=True)
print("\nVariance:\n", variance)

# Calcul des quantiles
quantiles = iris.quantile([0.25, 0.5, 0.75], numeric_only=True)
print("\nquantiles:\n", quantiles)
```

```python
# Création des boxplots univariés pour chaque variable numérique
fig, axes = plt.subplots(1, 4, figsize=(16, 6))

for i, column in enumerate(iris.columns[:-1]):
    sns.boxplot(y=column, data=iris, ax=axes[i])
    axes[i].set_title(f"Boxplot de {column}")

plt.tight_layout()
plt.show()
```

```python
# Visualisation des tendances centrales et de la dispersion à l'aide de boxplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for i, column in enumerate(iris.columns[:-1]):
    sns.boxplot(x="species", y=column, data=iris, ax=axes[i//2, i%2])

plt.tight_layout()
plt.show()
```
